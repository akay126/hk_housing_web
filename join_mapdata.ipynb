```python
## Cleaning Centa last transations
df_centa_last['Price'] = df_centa_last['Price'].astype(str)
df_centa_last ['Price']  = list(map(lambda x: float(x.replace('M','')),df_centa_last['Price']))
df_centa_last ['Price']  = [x*1000000 if x<100000 else x for x in df_centa_last['Price']]
```

```python
df_centa_last ['Saleable_Area'] = df_centa_last ['Saleable_Area'].replace('-',float('NaN'))
```

```python
df_centa_last  = df_centa_last.dropna()
df_centa_last['Saleable_Area'] = df_centa_last['Saleable_Area'].astype(int)
df_centa_last['Price_Per_Sqft'] = (df_centa_last['Price']/df_centa_last['Saleable_Area']).round()
df_centa_last.to_csv('centa_last_CLEANED.CSV', index = False)
```
